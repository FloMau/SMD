```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def LCG(Seed, n, a, b, m):
        x = [Seed]
        for i in range(0, n-1):
            x += [(a * x[-1] + b) % m] #x[-1] ist die letzte berechnete Zahl
        return x
```

```python
parameters = []
periode = []
for a in range(0, 25):
    random = LCG(0, 10000, a, 3, 1024)
    random2 = []
    j=0
    while random[j] not in random2:
        random2.append(random[j])
        j+=1
    k=0
    while random[k] != random[j]: #Weil nicht zwangsläufig die erste Zahl der oberen liste zuerst wieder auftaucht
        k+=1
    parameters.append(a)
    periode.append(j-k)
```

```python
plt.hist(LCG(0, 10000, 5, 3, 1024))
```

```python
plt.plot(parameters, periode, 'rx')
plt.xlabel('Parameterwert a')
plt.ylabel('Periodenlänge')
```
